# **PLAYGROUND RULES**
- *Don't commit experiments to git*:
    * While it looks otherwise, this is basically a JSON file: Metadata changes with every action. Most changes are pure noise to git.
    * If it is worth saving as a test, *save it as a test*.
    * If it is worth saving as part of project code, *save it as part of project code*.
    * If you want to break this rule, justify it in a comment over the experiment you want to commit.
    * It is fine to save this file, as long as you don't ***recklessly*** commit it to git.
    * Sadly, this file should not be .gitignored.
- When writing an experiment, clearly organize it in new, dedicated code blocks.
    * Remove it if it seems appropriate once you are done.
- Find project logic by importing from the ruter_chatbot namespace:
  ```python
  from ruter_chatbot.some_module import some_thing
  ```
  Pressing tab after the . should give you suggestions from existing modules.
  Pressing tab after the whitespace following import should give you suggestions within the current module.
